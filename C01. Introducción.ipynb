# ¿Qué es el Deep learning?

Inicialmente, en los años 50 del siglo pasado, se tuvieron los inicios de la Inteligencia Artificial. El objetivo de esta temprana IA era crear máquinas y sistemas automáticos que se comportaran como un ser humano. 

Mas tarde, a eso de los 80 comienza el Machine Learning, que es un subapartado de la Inteligencia Artificial enfocado en usar datos históricos y estadísticos para que la máquina aprenda a tomar decisiones y hacer predicciones. 

Finalmente, cercano al 2010 hay un despegue del Deep Learning enfocado fuertemente a las redes neuronales.

El Deep Learning ha transformado radicalmente la forma en que las máquinas interactúan con el mundo real. Su importancia radica en que permite a los modelos aprender representaciones complejas de los datos de manera automática, sin necesidad de diseñar manualmente características (features).

A diferencia de los enfoques clásicos, el Deep Learning:

- Aprende jerarquías de información: de lo simple a lo complejo.

- Escala muy bien cuando se dispone de grandes volúmenes de datos.

- Ha permitido resolver problemas que antes eran prácticamente intratables.

Gracias a esto, el Deep Learning es hoy el motor principal detrás de los avances recientes en Inteligencia Artificial, como asistentes virtuales, traducción automática, visión por computadora y modelos generativos.



## ¿Cómo funciona el Deep learning?

Se basa en el funcionamiento de las neurona. Así, las redes neuronales *artificiales* son representaciones matemáticas de las redes neuronales *biológicas*.
![imagenes](im02.png)

A su vez, estas redes se pueden volver muy complejas añadiendo *capas*, lo que da a lugar a lo que se conoce como una *red profunda*, y de ahí viene el nombre de Deep learnging

![imagenes](im03.png)

El hecho de tener varias capas permite a las redes neuronales generalizar muy bien. Al contrario que las técnicas clásicas de Machine Learning que pueden presentar problemas de sobrememorización en el entrenamiento, las redes neuronales toman las decisiones basadas no solo en los datos como tal, sino también en la misma estructura.

### Diferencia entre Machine Learning y Deep Learning

#### Machine Learning (ML)

El Machine Learning clásico se basa en:

- Modelos relativamente simples (regresión, árboles, SVM, k-NN).

- Ingeniería de características manual: el analista decide qué variables son importantes.

- Buen desempeño con pocos datos.

- Mayor interpretabilidad.

#### Deep Learning (DL)

El Deep Learning:

- Utiliza redes neuronales profundas (muchas capas).

- Aprende automáticamente las características relevantes.

- Requiere grandes cantidades de datos y poder computacional.

- Es menos interpretable, pero mucho más potente.

El Deep Learning es especialmente adecuado cuando:

- Los datos son no estructurados (imágenes, audio, texto).

- Hay muchos datos disponibles.

- El problema es altamente complejo.

- Se busca el mejor desempeño posible, aun sacrificando interpretabilidad.

- Para problemas pequeños o tabulares, Machine Learning clásico suele ser suficiente y más eficiente.

## Aplicaciones del Deep Learning

**1. Visión por computadora**

- Reconocimiento facial (desbloqueo de celulares, control de acceso).

- Diagnóstico médico a partir de imágenes (radiografías, tomografías, lesiones en la piel).

- Detección de objetos en vehículos autónomos (peatones, señales, otros autos).

Aquí, las redes profundas aprenden bordes y texturas en capas iniciales, formas y patrones intermedios y objetos completos en capas profundas.

**2. Procesamiento de lenguaje natural (NLP)**

- Traducción automática (Google Translate, DeepL).

- Chatbots y asistentes virtuales.

- Análisis de sentimiento en redes sociales.

- Resúmenes automáticos de textos.

Los modelos profundos permiten capturar contexto, ambigüedad del lenguaje y dependencias a largo plazo entre palabras.

**3. Audio y voz**

- Reconocimiento de voz (Siri, Alexa, Google Assistant).

- Conversión texto-a-voz y voz-a-texto.

- Identificación de hablantes.

Aquí el Deep Learning aprende patrones temporales complejos en señales de audio.

**4. Juegos y toma de decisiones**

- AlphaGo y AlphaZero.

- Sistemas que juegan videojuegos sin reglas explícitas.

- Optimización de estrategias en entornos complejos.

Estos modelos no solo reconocen patrones, sino que aprenden a decidir.

**5. Arte y creatividad**

- Generación de imágenes (arte digital, fotografía sintética).

- Música generada por IA.

- Restauración y colorización de imágenes antiguas.

El Deep Learning ha difuminado la frontera entre lo técnico y lo creativo.
